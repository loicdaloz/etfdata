```python
import yfinance as yf
import pandas as pd

# Load the ticker symbols from the CSV file
df = pd.read_csv('etf (3).csv')

# Create an empty DataFrame to store the historical prices
prices = pd.DataFrame()

# Loop over each ticker symbol in the CSV file
for ticker in df['TICKER']:
    # Use yfinance to retrieve the historical prices for the ticker
    ticker_prices = yf.download(ticker, start='2010-01-01', end='2023-03-23')
    
    # Add a column to the DataFrame with the ticker symbol
    ticker_prices['TICKER'] = ticker
    
    # Append the ticker's historical prices to the overall DataFrame
    prices = pd.concat([prices, ticker_prices])

# Save the historical prices to a new CSV file
prices.to_csv('historical_prices.csv')
```
